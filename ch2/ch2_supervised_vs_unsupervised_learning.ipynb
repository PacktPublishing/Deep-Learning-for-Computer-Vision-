## Visualizing Handwritten Digits from the MNIST Dataset

This code loads the MNIST digits dataset, which consists of 8x8 pixel images of handwritten digits, and visualizes the first ten images in a 2x5 grid. Each image is displayed in grayscale with its corresponding label shown above as a title. The `imshow` function from `matplotlib` is used to render the images, and `axis('off')` removes the axis ticks and labels for a cleaner look. The `tight_layout()` method adjusts the spacing between subplots for better presentation before displaying the plot with `plt.show()`.

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Plot the first few digits with their labels
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.imshow(X[i].reshape(8, 8), cmap='gray')
    ax.set_title(f'Label: {y[i]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Implementing Logistic Regression with Elastic Net Regularization

This code demonstrates the process of applying a Logistic Regression model with Elastic Net regularization to the MNIST dataset for digit classification. It starts by loading the dataset and then scales the features to standardize the data, improving the model's performance. The data is then split into training and test sets to evaluate the model's accuracy on unseen data. The Logistic Regression model is configured with Elastic Net regularization to combine L1 and L2 penalties, making it suitable for models where feature selection and regularization are necessary. After training the model on the scaled training data, predictions are made on the test set. Finally, the model's accuracy is evaluated by comparing the predicted digit labels against the actual labels, providing a quantitative measure of the model's performance in classifying handwritten digits.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the Logistic Regression model with Elastic Net regularization
log_reg_model = LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=0.5, C=1.0, max_iter=1000)

# Fit the model on the training data
log_reg_model.fit(X_train, y_train)

# Predict on the test set
y_pred = log_reg_model.predict(X_test)

# Calculate the accuracy
accuracy1 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy1}')
```

## Applying K-Nearest Neighbors Classifier for Classification

This code fragment illustrates the application of the K-Nearest Neighbors (KNN) algorithm, a simple yet effective machine learning model for classification tasks. The `KNeighborsClassifier` from `scikit-learn` is instantiated with `n_neighbors=5`, indicating that the model will consider the five nearest neighbors in its voting process for classifying new instances. The model is then trained on a dataset (`X_train`, `y_train`), where `X_train` represents the features of the training samples, and `y_train` contains the corresponding class labels. After the training phase, the model is used to predict the classes of a separate test dataset (`X_test`). The accuracy of these predictions, compared to the true labels (`y_test`), is evaluated using the `accuracy_score` function, providing a straightforward metric to assess the model's performance. This method showcases the KNN algorithm's capability to classify new data points based on the majority vote of their neighbors, emphasizing its utility in scenarios where the decision boundary is not linear.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale

# Load MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling
X_scaled = scale(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the KNeighbors Classifier with k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model on the training data
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate the accuracy
accuracy2 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy2}')
```

## Utilizing Decision Tree Classifier for Predictive Modeling

This code snippet outlines the process of employing a Decision Tree Classifier for classification purposes, showcasing its simplicity and effectiveness in making predictions. By instantiating the `DecisionTreeClassifier` from `scikit-learn` with a `max_depth` of 100, the model is allowed to grow complex enough to capture the nuances in the training data, potentially improving its ability to generalize to new, unseen data. The model is then trained on a designated training dataset (`X_train`, `y_train`), where `X_train` comprises the feature vectors and `y_train` the corresponding labels. Following the training phase, the model's predictive capability is tested on a separate dataset (`X_test`), and the predictions are compared to the actual labels (`y_test`) to calculate the model's accuracy. The final output is the accuracy score, which quantifies the model's performance by reflecting the proportion of correct predictions made out of all predictions. This demonstrates the decision tree's capacity to efficiently segment the feature space and make predictions, making it a valuable tool for various classification tasks.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling (optional for Decision Trees but included for consistency)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the Decision Tree classifier
decision_tree_model = DecisionTreeClassifier(max_depth=100)

# Fit the model on the training data
decision_tree_model.fit(X_train, y_train)

# Predict on the test set
y_pred = decision_tree_model.predict(X_test)

# Calculate the accuracy
accuracy3 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy3}')
```

## Implementing Support Vector Machine Classifier with RBF Kernel

This code snippet illustrates the application of a Support Vector Machine (SVM) classifier using a radial basis function (RBF) kernel to address classification tasks. The SVM model, created through the `SVC` class from `scikit-learn`, leverages the RBF kernel to handle non-linear relationships between the features by mapping input data into a higher-dimensional space where a linear separator can be identified. After initializing the SVM classifier, the model is trained with a set of labeled data (`X_train` and `y_train`), allowing it to learn the distinctions among different classes based on the training examples provided. Subsequently, the trained model is used to make predictions on a separate test dataset (`X_test`), and the accuracy of these predictions is evaluated against the actual labels (`y_test`) using the `accuracy_score` function. The accuracy metric, printed at the end, quantitatively assesses the classifier's performance, showcasing the SVM's effectiveness in classifying complex datasets by finding the optimal hyperplane that separates the different classes with the maximum margin.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the SVM classifier with a radial basis function (RBF) kernel
svm_model = SVC(kernel='rbf')

# Fit the model on the training data
svm_model.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_model.predict(X_test)

# Calculate the accuracy
accuracy4 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy4}')
```

## Leveraging Gradient Boosting for Enhanced Classification Accuracy

This code snippet demonstrates the use of a Gradient Boosting Classifier, a powerful ensemble technique that builds a predictive model in a stage-wise fashion. By instantiating `GradientBoostingClassifier` from `scikit-learn` with parameters specifying the number of boosting stages (`n_estimators=100`), the rate of learning (`learning_rate=0.1`), and the maximum depth of individual regression estimators (`max_depth=5`), the model is configured to iteratively correct errors of previous predictors with new ones, effectively improving its prediction accuracy over time. The model is trained using a labeled dataset (`X_train` and `y_train`), allowing it to learn from the training examples. After training, the model is applied to predict the labels of a test dataset (`X_test`), and the accuracy of these predictions is evaluated against the true labels (`y_test`) using the `accuracy_score` function. The resulting accuracy metric quantifies the model's performance, showcasing the effectiveness of Gradient Boosting in tackling complex classification problems by focusing on difficult cases and minimizing prediction errors through successive learning iterations.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the Gradient Boosting classifier
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5)

# Fit the model on the training data
gb_model.fit(X_train, y_train)

# Predict on the test set
y_pred = gb_model.predict(X_test)

# Calculate the accuracy
accuracy5 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy5}')
```

## Leveraging Random Forest Classifier for Enhanced Accuracy

This code snippet demonstrates the use of the Random Forest classifier, a powerful ensemble learning method that integrates multiple decision trees to improve classification accuracy and control over-fitting. By initializing the `RandomForestClassifier` from `scikit-learn` with `n_estimators=100`, it configures the ensemble to consist of 100 individual decision trees, each contributing to the final prediction. The `max_depth=None` parameter allows the trees to grow as deep as necessary to achieve high accuracy on the training data. The model is trained using a dataset (`X_train`, `y_train`), where `X_train` includes the features and `y_train` comprises the corresponding labels. Following training, the ensemble model is tasked with predicting the classes of a separate test set (`X_test`). The predictions are then compared to the actual test labels (`y_test`) to compute the model's accuracy, which quantifies how well the Random Forest classifier performs on unseen data. The output accuracy score reflects the ensemble method's effectiveness in leveraging the strengths of multiple trees, thereby reducing the likelihood of overfitting and enhancing the predictive performance on complex classification tasks.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data by scaling (optional for Random Forest but included for consistency)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, max_depth=None)

# Fit the model on the training data
rf_model.fit(X_train, y_train)

# Predict on the test set
y_pred = rf_model.predict(X_test)

# Calculate the accuracy
accuracy6 = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy6}')
```

# Famous Unsupervised Learning Algorithms

We will load the digits dataset and only use six first of the ten available classes.

```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=6)
X, y = digits.data, digits.target
n_samples, n_features = X.shape
n_neighbors = 30
```

### Helper function to plot embedding

Below, we will use different techniques to embed the digits dataset. We will plot the projection of the original data onto each embedding. It will allow us to check whether or digits are grouped together in the embedding space, or scattered across it.

*   List item
*   List item



```python
import numpy as np
from matplotlib import offsetbox

from sklearn.preprocessing import MinMaxScaler


def plot_embedding(X, title):
    _, ax = plt.subplots()
    X = MinMaxScaler().fit_transform(X)

    for digit in digits.target_names:
        ax.scatter(
            *X[y == digit].T,
            marker=f"${digit}$",
            s=60,
            color=plt.cm.Dark2(digit),
            alpha=0.425,
            zorder=2,
        )
    shown_images = np.array([[1.0, 1.0]])  # just something big
    for i in range(X.shape[0]):
        # plot every digit on the embedding
        # show an annotation box for a group of digits
        dist = np.sum((X[i] - shown_images) ** 2, 1)
        if np.min(dist) < 4e-3:
            # don't show points that are too close
            continue
        shown_images = np.concatenate([shown_images, [X[i]]], axis=0)
        imagebox = offsetbox.AnnotationBbox(
            offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r), X[i]
        )
        imagebox.set(zorder=1)
        ax.add_artist(imagebox)

    ax.set_title(title)
    ax.axis("off")
```

### Embedding techniques comparison

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomTreesEmbedding
from sklearn.manifold import (
    MDS,
    TSNE,
    Isomap,
    LocallyLinearEmbedding,
    SpectralEmbedding,
)
from sklearn.neighbors import NeighborhoodComponentsAnalysis
from sklearn.pipeline import make_pipeline
from sklearn.random_projection import SparseRandomProjection

embeddings = {
    "Random projection embedding": SparseRandomProjection(
        n_components=2, random_state=42
    ),
    "Truncated SVD embedding": TruncatedSVD(n_components=2),
    "Linear Discriminant Analysis embedding": LinearDiscriminantAnalysis(
        n_components=2
    ),
    "Isomap embedding": Isomap(n_neighbors=n_neighbors, n_components=2),
    "Standard LLE embedding": LocallyLinearEmbedding(
        n_neighbors=n_neighbors, n_components=2, method="standard"
    ),
    "Modified LLE embedding": LocallyLinearEmbedding(
        n_neighbors=n_neighbors, n_components=2, method="modified"
    ),
    "Hessian LLE embedding": LocallyLinearEmbedding(
        n_neighbors=n_neighbors, n_components=2, method="hessian"
    ),
    "LTSA LLE embedding": LocallyLinearEmbedding(
        n_neighbors=n_neighbors, n_components=2, method="ltsa"
    ),
    "MDS embedding": MDS(n_components=2, n_init=1, max_iter=120, n_jobs=2),
    "Random Trees embedding": make_pipeline(
        RandomTreesEmbedding(n_estimators=200, max_depth=5, random_state=0),
        TruncatedSVD(n_components=2),
    ),
    "Spectral embedding": SpectralEmbedding(
        n_components=2, random_state=0, eigen_solver="arpack"
    ),
    "t-SNE embedding": TSNE(
        n_components=2,
        n_iter=500,
        n_iter_without_progress=150,
        n_jobs=2,
        random_state=0,
    ),
    "NCA embedding": NeighborhoodComponentsAnalysis(
        n_components=2, init="pca", random_state=0
    ),
}
```

Once we declared all the methods of interest, we can run and perform the projection of the original data. We will store the projected data as well as the computational time needed to perform each projection.

```python
from time import time

projections, timing = {}, {}
for name, transformer in embeddings.items():
    if name.startswith("Linear Discriminant Analysis"):
        data = X.copy()
        data.flat[:: X.shape[1] + 1] += 0.01  # Make X invertible
    else:
        data = X

    print(f"Computing {name}...")
    start_time = time()
    projections[name] = transformer.fit_transform(data, y)
    timing[name] = time() - start_time
```

### Final Plot

Finally, we can plot the resulting projection given by each method.

```python
for name in timing:
    title = f"{name} (time {timing[name]:.3f}s)"
    plot_embedding(projections[name], title)

plt.show()
```
